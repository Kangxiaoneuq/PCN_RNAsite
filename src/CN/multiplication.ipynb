```python
import pandas as pd
import numpy as np
ori= pd.read_excel(r'CS71.xlsx')
data=ori.iloc[:142,0]

xl=[]#序列
fasta=[]#序列名字
xln=[]#每个序列的长度

for i in range(len(data)):
    if i%2==0:
        a=data.iloc[i]
        fasta.append(a)
    else:
        c=data.iloc[i]
        xl.append(c)
        xln.append(len(c))

# 把序列全部分开为字母
def Kmers_funct(seq, size): 
    return [seq[x:x+size] for x in range(len(seq) - size + 1)]
one_mers=[]
for i in range(len(xl)):
    myseq = xl[i]
    one_mers.append(Kmers_funct(myseq, size=1))

# 把所有连起来，成一个长序列
zimu=[]
for i in range(len(one_mers)):
    zimu.extend(one_mers[i])
    
#计算每一个字母的绝对位置
sl=[]
for i in range(len(xln)):
    a=1
    b=[]
    while a<=xln[i]:
        b.append(a/xln[i])
        a=a+1
    sl.append(b)
    
#把所有连起来，成一个长序列
SL=[]
for i in range(len(sl)):
    SL.extend(sl[i])
    
NPS = []
for i in range(len(xln)):
    if xln[i]%2 == 1:
        for j in range(int((xln[i]+1)/2),0,-1):
                NPS.append(j/((xln[i]+1)/2))
        for j in range(2,int((xln[i]+3)/2)):
                NPS.append(j/((xln[i]+1)/2))
    else:
        for j in range(int(xln[i]/2),0,-1):
            NPS.append(j/(xln[i]/2))
        for j in range(1,int((xln[i]+2)/2)):
            NPS.append(j/(xln[i]/2))
            
juzhen = ori.iloc[:,19:49]

#建立一个空白矩阵，随之往里填充
sl_juzhen = pd.DataFrame(np.arange(len(juzhen) * len(juzhen.columns)).reshape(len(juzhen),len(juzhen.columns)))
for i in range(len(juzhen.columns)):
    sl_juzhen.iloc[:,i] = juzhen.iloc[:,i] * SL

nps_juzhen = pd.DataFrame(np.arange(len(juzhen) * len(juzhen.columns)).reshape(len(juzhen),len(juzhen.columns)))
for i in range(len(juzhen.columns)):
    nps_juzhen.iloc[:,i] = juzhen.iloc[:,i] * NPS
    
result = pd.concat([sl_juzhen,nps_juzhen],ignore_index=True,axis=1) 

result.to_csv("CN2.csv", index=False )
```
